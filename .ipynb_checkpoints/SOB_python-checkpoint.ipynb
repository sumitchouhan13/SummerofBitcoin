```python
#Total values in the parents column
df['parents '].isna().sum()
```

```python
#First adding the nan values in the parent column or adding the rows which doesnt have parents in it
total_set = []
for index, row in df.iterrows():
    if(pd.isnull(row['parents '])):
        total_set.append(row)
    else:
        string = []
        string = row['parents '].split(';')
        for i in string:
            for j in total_set:
                if i == j['tx_id']:
                    row['fee'] = row['fee'] + j['fee']
                    row['weight'] = row['weight'] + j['weight']
        total_set.append(row)
```

```python
#Now deleting the unnecessary columns from the list created
for j in total_set:
    del j['parents ']
```

```python
feepweight = [] #This is fee per weight 
for index, row in df.iterrows():
    feepweight.append(row['fee']//row['weight'])
df['feepweight'] = feepweight
```

```python
df.drop(['parents '], axis = 1)
```

```python
test = df.sort_values(['feepweight'], ascending=[False])
test
```

```python
# Now as we know that the weight of the block can not exceed 4,000,000 weight so the problem breaks down to fractional knapsack
total_profit = 0
total_weight = 0
file1 = open("block.txt","w")
for index, row in test.iterrows():
    if total_weight + row['weight'] <= 4000000:
        total_weight = total_weight + row['weight']
        total_profit = total_profit + row['fee']
    else:
        remain = 4000000 - total_weight
        total_profit = (row['fee'] / row['weight']) * remain
        break
    file1.write(row['tx_id'] + "\n")
print("The Total Profit earned by the miner is :" , total_profit)
file1.close()
```
